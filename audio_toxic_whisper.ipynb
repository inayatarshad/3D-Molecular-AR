```python
# âœ… Install yt-dlp (for YouTube downloads)
!pip install -q yt-dlp

# âœ… Python script to download multiple videos as audio into Google Drive
import yt_dlp

# List of YouTube video URLs
urls = [
    "https://www.youtube.com/watch?v=PbSYN7jTWcc&pp=ygUcZHVja3kgYmhhaSByb2FzdHMgdW5maWx0ZXJlZA%3D%3D",
    "https://www.youtube.com/watch?v=ttXxJWplrkU&pp=ygUcZHVja3kgYmhhaSByb2FzdHMgdW5maWx0ZXJlZA%3D%3D",
    "https://www.youtube.com/watch?v=abhuAYtmk58&pp=ygUMY2FycnltaW5hdGkg0gcJCbIJAYcqIYzv",
    "https://www.youtube.com/watch?v=vcLRWiTNCbQ&pp=ygUMY2FycnltaW5hdGkg",
    "https://www.youtube.com/watch?v=bHLJjgzCDzU&pp=ygUMY2FycnltaW5hdGkg"
]

# Output folder in Google Drive
output_dir = "/content/drive/MyDrive/audio_toxic/%(title)s.%(ext)s"

# yt-dlp options
ydl_opts = {
    "format": "bestaudio/best",
    "outtmpl": output_dir,  # save in Google Drive
    "postprocessors": [{
        "key": "FFmpegExtractAudio",
        "preferredcodec": "mp3",
        "preferredquality": "192",
    }],
}

# Download all videos
with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download(urls)

print("âœ… Download completed! Files are saved in /content/drive/MyDrive/audio_toxic/")
```

```python
# âœ… Install Whisper + ffmpeg
!pip install -q openai-whisper
!sudo apt-get -y install ffmpeg

import whisper
import os

# Path to your audio folder in Google Drive
audio_dir = "/content/drive/MyDrive/audio_toxic"
output_dir = audio_dir  # transcripts will be saved in the same folder

# Load Whisper model (options: tiny, base, small, medium, large)
model = whisper.load_model("small")

# Loop through all MP3 files in folder
for file in os.listdir(audio_dir):
    if file.endswith(".mp3"):
        audio_path = os.path.join(audio_dir, file)
        print(f"ğŸ”Š Transcribing: {file} ...")

        # Run transcription
        result = model.transcribe(audio_path, language="ur")

        # Save transcript as .txt
        text_filename = os.path.splitext(file)[0] + ".txt"
        text_path = os.path.join(output_dir, text_filename)
        with open(text_path, "w", encoding="utf-8") as f:
            f.write(result["text"])

        print(f"âœ… Transcript saved: {text_path}")

print("ğŸ‰ All transcriptions completed!")
```
